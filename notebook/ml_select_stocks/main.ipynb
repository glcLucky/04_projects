```python
# 更新月度原始特征
# stock_pool = pd.read_csv(r"E:\07_data\02_factor\temp_data\zz500.csv")['wind_code'].tolist()  # zz500
start_lst = [str(2008+i) + '-{:0>2}'.format(str(1+j)) + '-01' for i in range(0, 13) for j in range(0, 12)]
end_lst = [str(2008+i) + '-{:0>2}'.format(str(1+j)) + '-31' for i in range(0, 13) for j in range(0, 12)]
# start_lst = ['2018-04-01', '2018-05-01', '2018-06-01', '2018-07-01','2018-08-01']
# end_lst = ['2018-04-31', '2018-05-31', '2018-06-31', '2018-07-31','2018-08-31']
for i in range(len(start_lst)):
    a = str(int(start_lst[i][:4]) - 1) + '-12-31' # 选取哪一天的股票池 上年最后一天
    print(a)
    b = os.path.join(root_path_pool, "{}_stocks_{}.json".format(stocks_pool_name, a))
    c = dk.json2dict(b)
    df_inds = alpha.api.get_monthly_feats_on_given_period(start_lst[i], end_lst[i], c)
    raw_inds_month_path = os.path.join(DB_PATH, r"datasets\raw_indicators_by_month")
    if not os.path.exists(raw_inds_month_path):
        os.makedirs(raw_inds_month_path)
    df_inds.to_csv(os.path.join(raw_inds_month_path, r"{}.csv".format(start_lst[i][:7])), index=0)
```

```python
start_lst = [str(2004+i) + '-{:0>2}'.format(str(1+j)) + '-01' for i in range(0, 15) for j in range(0, 12)]
end_lst = [str(2004+i) + '-{:0>2}'.format(str(1+j)) + '-31' for i in range(0, 15) for j in range(0, 12)]
back_test_point_st = '2013-03'  # 回测时间开始月份
back_test_point_end = '2018-08'  # 回测时间结束月份
# back_test_point_st 回测时间开始月份
# predict_secs_point 预测月份
# train_st_lst 训练区间
for i in range(0, 1000):  # START_DATE往后12个月
    back_test_point = dk.date2char(dk.char2datetime(back_test_point_st + '-01') +  dk.timedelta({'months': i}))[:7]
    if back_test_point > back_test_point_end:
        break
    print(back_test_point)
    predict_secs_point = dk.date2char(dk.char2datetime(back_test_point + '-01') -  dk.timedelta({'months': 1}))[:7]
    # 训练期 train_st_lst[0]往后60个月的数据 测试期： train_st_lst[0]往后第61个月数据
    train_st = dk.date2char(dk.char2datetime(predict_secs_point + '-01')  - dk.timedelta({'months': 61}))[:7] + '-01'
    start_lst_ix = start_lst.index(train_st)
    train_st_lst = start_lst[start_lst_ix: start_lst_ix+61]
    test_st = [start_lst[start_lst_ix+61]]
    print("训练集开始日期:", train_st_lst[0])
    print("训练集结束日期:", train_st_lst[-1])
    print("测试集日期:", test_st[0])
    dataID = train_st_lst[0][:7]  # 训练集起始年月
    df_train_dirty = alpha.api.get_analysis_data(date_list=train_st_lst, dataID=dataID, type="train", override=True)
    df_test_dirty = alpha.api.get_analysis_data(date_list=test_st, dataID=dataID, type="test", override=True)
    df_train_clean = alpha.api.preprocess_train_data(df_train_dirty, dataID, override=True)
    # 分离特征和标签¶
    X_train = df_train_clean.iloc[:, :-1]
    y_train = df_train_clean.iloc[:, -1]

    #训练xgboost模型
    params = {'max_depth': 2, 'eta': 0.05, 'silent': 1, 'objective': 'binary:logistic', 'eval_metric':'auc'}
    # num_boost_round必须放在xgb.train语句中 放在params中无效 并且该参数是一个重要参数
    bst = alpha.api.XGB_classfier_fit(X_train, y_train, params, train_st_lst[0][:7], 200, refit= True)

    # 测试集预处理
    X_test = alpha.api.preprocess_test_data(df_test_dirty, train_st_lst[0][:7], X_train.columns.tolist(), override=True)
    # X_test = df_test_clean.iloc[:,:-1]
    # y_test = df_test_clean.iloc[:,-1].to_frame()

    # 模型评估-统计视角
    # y_train_scores = alpha.api.XGB_classfier_predict_prob(bst, X_train)
    # y_test_scores = alpha.api.XGB_classfier_predict_prob(bst, X_test)
    # alpha.api.plot_roc_curve(y_train, y_train_scores, "训练集", train_st_lst[0])
    # alpha.api.plot_roc_curve(y_test, y_test_scores, "测试集", train_st_lst[0])

    # 模型评估-回测视角
    # 根据当前日期(月末)的特征计算选中的股票
    NUM = 5
    selected_stocks = [] # key为yearmonth value为股票列表 表示在该月份持有的股票
    X_now = X_test.copy() # 本月末的特征
    X_now['prob'] = alpha.api.XGB_classfier_predict_prob(bst, X_now) # 根据本月末的特征计算出来股票池中每只股票为强势股的概率
    X_now = X_now.sort_values(['prob'])
    selected_stocks = X_now.iloc[-NUM:, -1].reset_index()['sec_id'].tolist()
    yearmonth = dk.date2char(dk.char2datetime(test_st[0]) +  dk.timedelta({'months': 1}))[:7]
    
    #############  在此之前不需要测试当月的月底股价  ############
    alpha.api.backtest_for_mfs(yearmonth, selected_stocks, "000905.SH", "XGBoost", "label_A")
```
