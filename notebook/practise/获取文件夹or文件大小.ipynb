```python
import os
import pandas as pd
```

# 获取文件大小

```python
file_path = r"C:\Users\youlin\AppData\Local\Apple\ie-dav.zm"
def get_file_size(file_path):
    """
    获取某个文件的大小（单位：M）
    """
    return os.path.getsize(file_path)/1024/1024
```

# 获取某个文件夹大小

```python
folder_path = r"C:\Users\youlin\AppData\Local\Apple"
def  get_folder_size(folder_path):
    """
    获取某个文件夹大小（单位：M）
    """
    size = 0
    for root, dirs, files in os.walk(folder_path):
        size += sum([get_file_size(os.path.join(root, file)) for file in files])
    return size
        
```

# 获取某个文件夹内部各文件夹大小排名

```python
def  get_folder_size_str(folder_path):
    """
    获取某个文件夹大小（单位：M）
    """
    size = {}
    num = len(folder_path)
    for root, dirs, files in os.walk(folder_path):
        size_sum = 0
        for file in files:
            if file != 'MicrosoftEdge.exe':
                path = os.path.join(root, file)
                if os.path.exists(path):
                    size_sum += get_file_size(path)
        
        size[root[num+1:]] = size_sum
    a = pd.DataFrame([[k,v] for k,v in size.items()], columns=['name', 'size'])
    return a.sort_values(['size'], ascending=False)
```
